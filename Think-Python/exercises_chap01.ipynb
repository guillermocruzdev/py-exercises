### Exercise

The following questions give you a chance to practice writing arithmetic expressions.

1.  How many seconds are there in 42 minutes 42 seconds?


```python
total_seg = (42 * 60) + 42
print(f'{total_seg} seconds')
```


2.  How many miles are there in 10 kilometers? Hint: there are 1.61 kilometers in a mile.


```python
total_miles = 10 / 1.61
print(f'{total_miles:.2f} miles')
```


3.  If you run a 10 kilometer race in 42 minutes 42 seconds, what is your average pace in seconds per mile? 


```python
average_seg = total_seg / total_miles
print(f'{average_seg} seconds x mile')
```

    
4.  What is your average pace in minutes and seconds per mile?


```python
average_min = average_seg // 60
seconds = average_seg - 360
print(f'{int(average_min)} minutes and {seconds:.2f} seconds x mile')
```


5.  What is your average speed in miles per hour?



```python
total_hour = total_seg / 3600
average_speed = total_miles / total_hour

print (f'{average_speed:.2f} mph')
```
