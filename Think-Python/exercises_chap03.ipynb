#### Función `repeat` en Python

Esta función utiliza un bucle `for` y la función `range()` para imprimir el texto el número de veces especificado por el usuario.

```python
def repeat(string, n):
    """
    Imprime el string dado 'n' veces.
    """
    for i in range(n):
        print(string)

# Ejemplo de uso:
# repeat("Hola", 3)
```

#### Versión de `repeat` sin el bucle `for`

En Python, puedes evitar el uso de un bucle `for` aprovechando el **operador de repetición de cadenas** (`*`). Este operador permite multiplicar un string por un entero, creando una nueva cadena con el contenido repetido.

Para que cada repetición aparezca en una línea nueva (simulando el comportamiento del `print` dentro de un bucle), debemos añadir un carácter de salto de línea `\n`.

```python
def repeat(string, n):
    """
    Imprime el string 'n' veces sin usar un bucle for.
    """
    # Multiplicamos el string (con un salto de línea) por n
    # .rstrip() elimina el último salto de línea sobrante
    print((string + "\n") * n, end="")

# Ejemplo de uso:
# repeat("Hola", 3)
```

#### Error en la función `print_twice`

El problema principal en esta versión de la función es un **error de referencia de variable** (conocido en Python como `NameError`).

##### El Problema
En la definición de la función, has nombrado al parámetro de entrada como `string`, pero dentro del cuerpo de la función intentas imprimir una variable llamada `cat`.

* **Parámetro definido:** `string`
* **Variable utilizada:** `cat`

Como la variable `cat` no ha sido definida dentro de la función (ni se pasó como argumento), Python no sabrá qué valor mostrar y lanzará un error al intentar ejecutarla.

### Exercise

Write a function named `print_right` that takes a string named `text` as a parameter and prints the string with enough leading spaces that the last letter of the string is in the 40th column of the display.

````python
def print_right(text):
    total_width = 40
    spaces = total_width - len(text)
    print(" " * spaces + text)

print_right("Hello, World!")
````

Hint: Use the `len` function, the string concatenation operator (`+`) and the string repetition operator (`*`).

Here's an example that shows how it should work.

````python
print_right("Monty")
print_right("Python's")
print_right("Flying Circus")
````

### Exercise

Write a function called `triangle` that takes a string and an integer and draws a pyramid with the given height, made up using copies of the string. Here's an example of a pyramid with `5` levels, using the string `'L'`.

````python
def triangle(s, n):
    width = 2 * n - 1
    for i in range(n):
        row = s * (2 * i + 1)
        print(row.center(width))
````

````python
triangle('L', 5)
````

### Exercise

Write a function called `rectangle` that takes a string and two integers and draws a rectangle with the given width and height, made up using copies of the string. Here's an example of a rectangle with width `5` and height `4`, made up of the string `'H'`.

````python
def rectangle(s, width, height):
    for i in range(height):
        print(s * width)
````

````python
rectangle('H', 5, 4)
````

### Exercise

The song "99 Bottles of Beer" starts with this verse:

> 99 bottles of beer on the wall  
> 99 bottles of beer  
> Take one down, pass it around  
> 98 bottles of beer on the wall  

Then the second verse is the same, except that it starts with 98 bottles and ends with 97. The song continues -- for a very long time -- until there are 0 bottles of beer.

Write a function called `bottle_verse` that takes a number as a parameter and displays the verse that starts with the given number of bottles.

Hint: Consider starting with a function that can print the first, second, or last line of the verse, and then use it to write `bottle_verse`.

````python
def bottle_verse(n):
    print(f"{n} bottles of beer on the wall,")
    print(f"{n} bottles of beer,")
    print(f"Take one down, pass it around,")
````

Use this function call to display the first verse.

````python
bottle_verse(99)
````

If you want to print the whole song, you can use this `for` loop, which counts down from `99` to `1`.
You don't have to completely understand this example---we'll learn more about `for` loops and the `range` function later.

````python
for n in range(99, 0, -1):
    bottle_verse(n)
    print()
````
